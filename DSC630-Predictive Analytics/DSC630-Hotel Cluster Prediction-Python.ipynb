```python
# import linear discriminant analysis for possible feature reduction
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

# Create arrays for the features and target variable
y = merged['hotel_cluster'].values
X = merged.drop('hotel_cluster', axis=1).values

# perform lda to reduce number of components to 5
lda = LinearDiscriminantAnalysis(n_components=5)

# fit training model
X_lda = lda.fit(X, y).transform(X)

# print results
print("Original number of features:", X.shape[1])
print("Reduced number of features:", X_lda.shape[1])

# display explained variance ratios
print(lda.explained_variance_ratio_)

```

```python
# LOGISTIC REGRESSION PREDICTION

# Import the necessary modules
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# create arrays for train and test sets
y = merged['hotel_cluster'].values
X = merged.drop('hotel_cluster', axis=1).values

# standardize features
scaler = StandardScaler()
X_standardized = scaler.fit_transform(X)

# Create training and test sets
X_train, X_test, y_train, y_test = train_test_split(X_standardized, y, test_size = 0.2, random_state=42, stratify=y)


# Create the classifier: logreg
logreg = LogisticRegression(random_state=42, multi_class='ovr')

# Fit the classifier to the training data
logreg.fit(X_train, y_train)

# Predict the labels of the test set: y_pred
y_pred = logreg.predict(X_test)

# Print the accuracy
print(logreg.score(X_test, y_test))

# Compute and print the classification report
print(classification_report(y_test, y_pred))


# create array for Final test set
X_new = test_merge.values

# Predict and print the label for the the final test set
new_pred = logreg.predict(X_new)
print("Final Prediction: {}".format(new_pred))

# Predict and print the probabilities for the the final test set
# new_prob = logreg.predict_proba(X_new)
# print("Final Prediction: {}".format(new_prob))

# save results to daraframes
log_pred = pd.DataFrame(new_pred)
# log_prob = pd.DataFrame(new_prob)
```
