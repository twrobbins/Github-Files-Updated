```python
import numpy as np
import scipy.stats as stats
import pandas as pd

skew_scipy = stats.skew(business['stars'])
skew_pandas = pd.DataFrame(business['stars']).skew()[0]

print('Skew:', skew_scipy, skew_pandas)

kurtosis_scipy = stats.kurtosis(business['stars'])
kurtosis_pandas = pd.DataFrame(business['stars']).kurtosis()[0]

print('Kurtosis:', kurtosis_scipy, kurtosis_pandas)
```

```python
# create new dataframe to convert to dummies

# create dataframe with attributes
# good way to filter in python using regex
attributes = data.filter(regex='attributes')

print(attributes.head())
type(attributes)
```

```python
# data['price_range2'] = int(data['price_range'])

attr_dummies = pd.get_dummies(attributes)

attr_dummies.head()
# dummies.head()
```

```python
# check for correlations between attributes
# LATER DO PCA OR LDA FOR FEATURE REDUCTION OF HIGHLY CORRELATED VARIABLES
print(attr_dummies.corr())


# data_cor.corr()
print(len(attr_dummies))
print(len(data))
```

```python
# identify correlations with stars variable

# import libraries
import pandas as pd

# use corr_with function from pandas package
corr_test = attr_dummies.corrwith(data['stars']) 

corr_test
```
