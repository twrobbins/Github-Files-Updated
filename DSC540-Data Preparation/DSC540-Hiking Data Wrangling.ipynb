```python
# import reader
from csv import reader

# read data files 
data_rdr = reader(open('DSC540-AllTrails data - nationalpark.csv', 'rb'))
header_rdr = reader(open('DSC540-nationalpark headers.csv', 'rb'))

# write object to new list
data_rows = [d for d in data_rdr]
# use list comprehension to remove mismatched headers
header_rows = [h for h in header_rdr if h[0] in data_rows[0]]

# make list of all short headers
all_short_headers = [h[0] for h in header_rows]

# initialize lists
skip_index = []
final_header_rows = []

# iterate over headers in dataset
for header in data_rows[0]:
    # single out headers that don't match
    if header not in all_short_headers:
        # return indexes we should skip
        index = data_rows[0].index(header)
        skip_index.append(index)
    else:
        for head in header_rows:
            if head[0] == header:
                final_header_rows.append(head)
                break
 

# create list for new data
new_data = []

# include only data rows
for row in data_rows[1:]:
    # initialize list
    new_row = []
    # use enumerate to isolate indexes of data rows to skip
    for i, d in enumerate(row):
        # make sure index is not in skip list
        if i not in skip_index:
            new_row.append(d)
    # add entry to new_data list
    new_data.append(new_row)

# initialize list
zipped_data = []

# iterate through new_data list
for drow in new_data:
    # zip each row and add it to new array
    zipped_data.append(zip(final_header_rows, drow))
    
print(zipped_data[0:1])
```
